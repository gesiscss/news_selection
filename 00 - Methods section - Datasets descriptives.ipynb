# Read data

```python
import pandas as pd
import numpy as np

# read news results dataset
dataset_A = pd.read_csv(f'data/dataset_A.csv')

# read the all news domain visits dataset
dataset_B = pd.read_csv(f'data/dataset_B.csv')

# read Google associated dataset
dataset_C = pd.read_csv(f'data/dataset_C.csv')

# read Google independent dataset
dataset_D = pd.read_csv(f'data/dataset_D.csv')

# read non news visits
dataset_E = pd.read_csv(f'data/dataset_E.csv')

# read all domain visits (news and not newws)
dataset_U = pd.read_csv('data/dataset_U.csv')

# include only the participants that were part of the statistical test
set_participants = set(dataset_A[dataset_A['is_selected_result']]['participant'])
```

# Methods Section: Dataset descriptives

```python
# visits per participant
print(f'\nDataset U - Visits per participants (N={len(dataset_U)}):')
print(dataset_U['participant'].value_counts().describe())

# search pages per participant
print(f'\nDataset A - Search pages per participant (N={len(dataset_A.drop_duplicates("search_page"))}):')
print(dataset_A.drop_duplicates('search_page')['participant'].value_counts().describe())
print(dataset_A['is_analysis_period'].replace({
    True: 'After cut-off date:',
    False: 'Before cut-off date:',
}).value_counts())

# news visits per participant
print(f'\nDataset B - News visits per participant (N={len(dataset_B)}):')
print(dataset_B['participant'].value_counts().describe())
print(dataset_B['is_analysis_period'].replace({
    True: 'After cut-off date:',
    False: 'Before cut-off date:',
}).value_counts())


# news visits per participant
print(f'\nDataset C - Google associated news visits per participant (N={len(dataset_C)}):')
print(dataset_C['participant'].value_counts().describe())
print(dataset_C['is_analysis_period'].replace({
    True: 'After cut-off date:',
    False: 'Before cut-off date:',
}).value_counts())

# Google indepependant news visits per participant
print(f'\nDataset D - Google independent news visits per participant (N={len(dataset_D)}):')
print(dataset_D['participant'].value_counts().describe())
print(dataset_D['is_analysis_period'].replace({
    True: 'After cut-off date:',
    False: 'Before cut-off date:',
}).value_counts())

# non-news visits per participant
print(f'\nDataset E - Non news visits per participant (N={len(dataset_E)}):')
print(dataset_E['participant'].value_counts().describe())

```
